```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load (Remember to Change These)
mainpath="/Users/prophet60091/Documents/GitHub/pandas-challenge/Resources"
filename1="schools_complete.csv"
filename2="students_complete.csv"

fullpath1=os.path.join(mainpath,filename1)
fullpath2=os.path.join(mainpath,filename2)

# Read School and Student Data File and store into Pandas Data Frames
school = pd.read_csv(fullpath1)
student = pd.read_csv(fullpath2)

# Combine the data into a single dataset
cs= pd.merge(student, school, how="left", on=["school_name", "school_name"])

cs
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
students=cs['Student ID'].nunique()
schools=cs['school_name'].nunique()

tbudget=school['budget'].sum()
mavg=cs['math_score'].mean()
ravg=cs['reading_score'].mean()

pass_rate=(cs['math_score'].mean()+cs['reading_score'].mean())/2

math_pass=cs.loc[cs['math_score'] >= 70, :]['Student ID'].count()
mpr=(math_pass/students)*100

read_pass=cs.loc[cs['reading_score'] >= 70, :]['Student ID'].count()
rpr=(read_pass/students)*100

opr = (mpr + rpr) / 2
```

```python
summary1 = pd.DataFrame(
        {
            'Total Students' : students,
            'Total Schools' : schools,
            'Total Budget' : tbudget,
            'Average Math Score' : mavg,
            'Average Reading Score' : ravg,
            '% Passing Math' : mpr,
            '% Passing Reading' : rpr,
            '% Overall Passing Rate' : opr
        },
    index = [0]
)
summary1
```

```python
summary1.dtypes
```

```python
flot={'Total Students': float,'Total Schools': float,'Total Budget': float,'Average Math Score': float,'Average Reading Score': float,'% Passing Math': float,'% Passing Reading': float,'% Overall Passing Rate': float}
summary1 = summary1.astype(flot)
summary1.dtypes


```

```python
summary1["Total Students"] = summary1["Total Students"].map("{:,}".format)
summary1["Total Budget"] = summary1["Total Budget"].map("${:,}".format)
summary1["Average Math Score"] = summary1["Average Math Score"].map("{:.2f}%".format)
summary1["Average Reading Score"] = summary1["Average Reading Score"].map("{:.2f}%".format)
summary1["% Passing Math"] = summary1["% Passing Math"].map("{:.2f}%".format)
summary1["% Passing Reading"] = summary1["% Passing Reading"].map("{:.2f}%".format)
summary1["% Overall Passing Rate"] = summary1["% Overall Passing Rate"].map("{:.2f}%".format)

summary1.head()

```

```python
cs.head()
```

```python
studentsps = cs.groupby('school_name')['Student ID'].count()
studentsps
```

```python
totalbps = cs.groupby('school_name')['budget'].sum()
totalbps
```

```python
amss = cs.groupby('school_name')['math_score'].sum() / studentsps
amss
```

```python
avgrss=cs.groupby('school_name')['reading_score'].sum() / studentsps
avgrss
```

```python
pms = cs.loc[cs['math_score'] > 70, :].groupby('school_name')['math_score'].count()
prs = cs.loc[cs['reading_score'] > 70, :].groupby('school_name')['reading_score'].count()
pms,prs
```

```python
addendum = pd.concat([studentsps, 
                        totalbps, 
                        amss, 
                        avgrss, 
                        pms, 
                        prs], 
                        axis = 1)

addendum.reset_index(inplace=True)

addendum = addendum.rename(columns={'index' : 'school_name',  
                                              'Student ID' : 'Total Students',
                                              'budget' : 'Total School Budget',
                                                0 : 'Average Math Score',
                                                1 : 'Average Reading Score',
                                                'math_score' : 'passing math raw',
                                                'reading_score' : 'passing reading raw'
                                                })

addendum=addendum.set_index('school_name')
addendum.head(5)
```

```python

addendum = school.merge(addendum, on='school_name')

addendum['Per Student Budget'] = addendum['Total School Budget'] / addendum['Total Students']
addendum['% Passing Math'] = (addendum['passing math raw'] / addendum['Total Students']) * 100
addendum['% Passing Reading'] = (addendum['passing reading raw'] / addendum['Total Students']) * 100
addendum['% Overall Passing Rate'] = (addendum['% Passing Math'] + addendum['% Passing Reading']) / 2

addendum.reset_index(inplace=True)
addendum=addendum.set_index('school_name')

addendum
```

```python
addendum=addendum.drop(columns=['index', 'School ID','size','budget','passing math raw','passing reading raw'])
addendumu=addendum
addendum
```

```python
convert={'Total Students': float,
         'Total School Budget': float,
         'Average Math Score': float,
         'Average Reading Score': float,
         'Per Student Budget': float,
         '% Passing Math': float,
         '% Passing Reading': float,
         '% Overall Passing Rate': float}
addendum = addendum.astype(convert)
```

## Math Scores by Grade

```python
ninth = student.loc[student["grade"] == "9th"].groupby("school_name", as_index=False)
tenth = student.loc[student["grade"] == "10th"].groupby("school_name", as_index=False)
eleventh = student.loc[student["grade"] == "11th"].groupby("school_name", as_index=False)
twelveth = student.loc[student["grade"] == "12th"].groupby("school_name", as_index=False)
```

```python
nAvg = pd.DataFrame(ninth["math_score"].mean())
tAvg = pd.DataFrame(tenth["math_score"].mean())
eAvg = pd.DataFrame(eleventh["math_score"].mean())
twAvg = pd.DataFrame(twelveth["math_score"].mean())
```

```python
mathByGrade = pd.merge(nAvg, tAvg, on="school_name")
mathByGrade = pd.merge(mathByGrade, eAvg, on="school_name")
mathByGrade = pd.merge(mathByGrade, twAvg, on="school_name")
mathByGrade.columns = ["school_name","9th","10th","11th","12th"]
mathByGrade.head(15)
```

```python
mathByGrade.reset_index(inplace=True)
mathByGrade=mathByGrade.set_index('school_name')
mathByGrade=mathByGrade.drop(columns=['index'])
mathByGrade
```
